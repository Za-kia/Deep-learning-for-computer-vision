```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114]).astype(int)
```

```python
img2 = plt.imread('./images_2/tshirt.jpg')
tshirt = rgb2gray(img2)
```

```python
tshirt.shape
```

```python
plt.imshow(tshirt, cmap = plt.get_cmap('gray'))
plt.show()
```
