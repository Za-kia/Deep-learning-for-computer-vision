### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">نمایش گرادیان یک تصویر با تغییرات روشنایی فازی</div>

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('images/gradient.png',0)

kernel = np.array([[0,0,0], 
                   [-1/2,0,1/2], 
                   [0,0,0]])


# Extract Sobl Edges
edges = cv2.filter2D(image, -1 , kernel)

plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(edges, cmap='gray');plt.title("Edges");

print_min_and_max(image)
print_min_and_max(edges)
````

### <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">نمایش تغییرات از راست به چپ و از چپ به راست</div>

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('images/little_girl.jpg',0)

kernel1 = np.array([[0,0,0], 
                    [-1/2,0,1/2], 
                    [0,0,0]])

kernel2 = np.array([[0,0,0], 
                    [1/2,0,-1/2], 
                    [0,0,0]])

# Extract Sobl Edges
edges1 = cv2.filter2D(image, -1, kernel1)
edges2 = cv2.filter2D(image, -1, kernel2)
all_vertical_edges = cv2.add(edges1, edges2)

all_vertical_edges = 255 - all_vertical_edges

all_vertical_edges[all_vertical_edges<230]=0
all_vertical_edges[all_vertical_edges>230]=255

plt.figure(figsize=[15,7])
plt.subplot(231);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(234);plt.imshow(edges1, cmap='gray');plt.title("Edges 1");
plt.subplot(235);plt.imshow(edges2, cmap='gray');plt.title("Edges 2");
plt.subplot(236);plt.imshow(all_vertical_edges, cmap='gray');plt.title("Final edges");
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('images/little_girl.jpg',0)

kernel1 = np.array([[0,0,0], 
                    [-1/2,0,1/2], 
                    [0,0,0]])

kernel2 = np.array([[0,0,0], 
                    [1/2,0,-1/2], 
                    [0,0,0]])

# Extract Sobl Edges
edges1 = cv2.filter2D(image, -1, kernel1)
edges2 = cv2.filter2D(image, -1, kernel2)
all_vertical_edges = cv2.add(edges1, edges2)


plt.figure(figsize=[15,7])
plt.subplot(231);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(234);plt.imshow(edges1, cmap='gray');plt.title("Edges 1");
plt.subplot(235);plt.imshow(edges2, cmap='gray');plt.title("Edges 2");
plt.subplot(236);plt.imshow(all_vertical_edges, cmap='gray');plt.title("Final edges");
````
